```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
def system1(n_steps=1000):
    """Simulate a 3-dimensional with two slow correlated degrees of freedom,
    and one fast low-variance coordinate
    
    Returns
    -------
    x : np.array, shape=(n_steps, 3)
    """
    import simtk.openmm as mm
    system = mm.System()
    system.addParticle(1)
    force = mm.CustomExternalForce('10*(x^2 + y^2 - 1.5*x*y) + 100*(z^2)')
    force.addParticle(0, [])
    system.addForce(force)
    integrator = mm.LangevinIntegrator(1000, 100, 0.05)
    context = mm.Context(system, integrator)
    context.setPositions([[0,0,0]])
    context.setVelocitiesToTemperature(1000)
    
    x = np.zeros((n_steps, 3))
    for i in range(n_steps):
        x[i] = context.getState(getPositions=True).getPositions(asNumpy=True)._value
        integrator.step(1)
    return x
```

```python
x = system1(10000)
print x.shape

ylabels = ['dim1', 'dim2', 'dim3']
for i in range(3):
    subplot(3, 1, i+1)
    plot(x[:, i])
    ylabel(ylabels[i])
xlabel('Time')
show()
```

```python
plot(x[:,0], x[:,1])
xlabel('dim1'); ylabel('dim2')
show()
```
